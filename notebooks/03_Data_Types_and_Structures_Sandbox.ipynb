```python
# 3. Tuple unpacking and swapping

trade = ("APPL", 150, 10)  # (symbol, price, quantity)
symbol, price, quantity = trade

symbol, price * quantity
```

```python
# 4. Unique attendess and overlaps

day1 = ["Ada", "Alan", "Ada", "Grace"]
day2 = ["Grace", "Linus", "Ada"]

unique_attendees = set(day1) | set(day2)
overlapping_attendees = set(day1) & set(day2)
came_only_day1 = set(day1) - set(day2)

unique_attendees, overlapping_attendees, came_only_day1
```

```python
# 5. Diccionary by comprehension

words = ["cat", "giraffe", "ant", "elephant"]
{word: len(word) for word in words if len(word) >= 4}
```
