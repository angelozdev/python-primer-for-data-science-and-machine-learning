```python
# 1. FizzBuzz with loop-else

for i in range(1, 31):
    if i % 15 == 0:
        print("FizzBuzz")
    elif i % 5 == 0:
        print("Buzz")
    elif i % 3 == 0:
        print("Fizz")
    else:
        continue
```

```python
# 2. First Event

nums = range(1, 100)

for num in nums:
    if num % 2 == 0:
        print(f"First even number found: {num}")
        break
else:
    print("No even number found in the range.")
    print("none")
```

```python
# 3. Safe division


def safe_divide(a, b):
    try:
        result = a / b
    except ZeroDivisionError:
        return None
    else:
        return result


print(safe_divide(10, 2))
print(safe_divide(10, 0))
```

```python
# 4. Packing slips

quantities = [2, 5, 1, 8, 3]
skus = ["A123", "B456", "C789", "D012", "E345"]

for index in enumerate(zip(skus, quantities), start=1):
    slip_number, (sku, quantity) = index
    print(f"Slip {slip_number}: SKU {sku}, Quantity {quantity}")
```

```python
# 5. Pattern matching

routes = [("GET", "/users/42"), ("POST", "/login")]


def get_route_name(route):
    method, path = route
    match (method, path):
        case ("GET", "/users/42"):
            return "users"
        case ("POST", "/login"):
            return "login"
        case _:
            print("Unknown route")


for route in routes:
    print(route)
    print(get_route_name(route))
```

```python
# 6. Sum until limit

nums = range(1, 1000)


def sum_until_limit(numbers, limit):
    total = 0
    for num in numbers:
        print(f"Adding {num}, current total: {total}")
        total += num
        if total > limit:
            break
    return total


print(sum_until_limit(nums, 100))
```
