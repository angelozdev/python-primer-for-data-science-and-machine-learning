```python
from contextlib import contextmanager


@contextmanager
def open_file(file_path, mode):
    file = open(file_path, mode)
    try:
        yield file
    finally:
        file.close()


with open_file("INDEX.md", "r") as f:
    content = f.read()
    print(content)
```
